```python
df.drop('Year_Built',axis=1,inplace=True)
```

```python
#splitting the data for training and testing
#input data
x=df.drop('House_Price',axis=1)  #labelling the data
y=df['House_Price']

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) #20%=0.2
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#model implementation

from sklearn.linear_model import LinearRegression  
# LinearRegression is a class, to use a class first we have to create object

linear_regressor= LinearRegression()
#whatever the objct name we give it is model(linear_regressor)

#training model
linear_regressor.fit(x_train,y_train)
```

```python
#testing model(predicting results)
#predicts 30% of x_test for testing

y_pred_lr= linear_regressor.predict(x_test)
y_pred_lr
```

```python
y_test
```

```python
#model evaluation
#checking results based on y_test

print("Accuracy of LR on training set:",linear_regressor.score(x_train,y_train))
print("Accuracy of LR on test set:",linear_regressor.score(x_test,y_test))
```

```python
#checking for error (checking wrong prediction)
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print("---Linear Regression-Model evaluation---")
print("Mean Absolute Error(MAE):",mae(y_test,y_pred_lr))
print("Mean Squared Error(MSE):",mse(y_test,y_pred_lr))
print("Root Mean Squared Error(RMSE):",np.sqrt(mse(y_test,y_pred_lr)))
```

```python
def predict_price(Square_Footage=1562,Num_Bedrooms =3,Num_Bathrooms=2,Lot_Size=0.524058,Garage_Size=1,Neighborhood_Quality=10):
    temp_array=list()
    #Square_Footage
    temp_array=temp_array +[Square_Footage]
    
    #Num_Bedrooms
    temp_array=temp_array +[Num_Bedrooms]

    #Num_Bathrooms
    temp_array=temp_array +[Num_Bathrooms]

     #Lot_Size
    temp_array=temp_array +[Lot_Size]

     #Garage_Size
    temp_array=temp_array +[Garage_Size]

     #Neighborhood_Quality
    temp_array=temp_array +[Neighborhood_Quality]


    #converting into numpy array
    temp_array= np.array([temp_array])
    print(temp_array)

    #prediction
    return int(linear_regressor.predict(temp_array))
```

```python
final_price=predict_price(Square_Footage=1562,Num_Bedrooms =3,Num_Bathrooms=2,Lot_Size=0.524058,Garage_Size=1,Neighborhood_Quality=10)
print("the final prediction price(range):{} to {}".format(final_price-1000,final_price+500))
```
